```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
matrix = np.array([[56259, 10696,  3266],
 [ 3651, 51616,  1124],
 [  288,   431, 55984]])
```

```python
matrix
```

```python
np.sum(matrix, axis=1)
```

```python
matrix.astype('float') / np.sum(matrix, axis=0)
```

```python
fig = plt.figure(figsize=(12, 6))
sns.heatmap(matrix.astype('float') / np.sum(matrix, axis=1), annot=True,
            fmt='.2%', cmap='Blues', xticklabels=['Angry', 'Disappointed', 'Happy'],
            yticklabels=['Angry', 'Disappointed', 'Happy'])
plt.title("Confusion Matrix")
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```
